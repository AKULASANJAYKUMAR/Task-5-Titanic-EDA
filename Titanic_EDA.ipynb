```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("/Users/sanjaykumarakula/Downloads/Task 7/train.csv")   # or give full path
df.head()
```

```python
df.shape
df.info()
df.describe(include="all").T
```

```python
df.isnull().sum()
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

```python
sns.countplot(x=df['Survived'])
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x=df['Sex'])
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x=df['Pclass'])
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x=df['Embarked'])
plt.title("Embarked Port Distribution")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['Survived','Age','Fare']].dropna(), diag_kind='kde')
plt.suptitle("Pairplot: Survived vs Age vs Fare", y=1.02)
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare vs Survival")
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival by Port of Embarkation")
plt.show()
```
